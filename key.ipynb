```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# import torch
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
```

```python
linear_data = pd.read_csv('data_linear.csv')
```

```python
linear_data.columns = ['dien tich','gia']
```

```python
linear_data.head()
```

```python
# .iloc <=> Extract Data from DataFrame to Series
# values <=> Convert data from Series to 1D Numpy array
# reshape(-1,1) <=> Convert 1D Numpy Array to 2D numpy array, 1: number of column, -1: number of row (automatic calculate from number of elements)
X_linear = linear_data.iloc[:,0].values.reshape(-1,1)
y_linear = linear_data.iloc[:,1].values.reshape(-1,1)
```

```python
# Normalize Data
X_linear=MinMaxScaler().fit_transform(X_linear)
y_linear=MinMaxScaler().fit_transform(y_linear)
```

```python
#Visualize Scaled Data
plt.scatter(X_linear, y_linear)
plt.title('Linear Data Scaled')
plt.xlabel('m√©t vu√¥ng/100')
plt.ylabel('gi√°/100')
```

```python
#G·ª£i √Ω:
#- Vi·∫øt code t√≠nh h√†m loss function
#- Vi·∫øt code t√≠nh gradient descent
#- Vi·∫øt code t√≠nh gi√° tr·ªã model d·ª± ƒëo√°n ƒë∆∞·ª£c
# ....
# C√≥ th·ªÉ tham kh·∫£o t·∫°i Assignment HW1 ƒë√£ l√†m 

model = LinearRegression()
model.fit(X_linear,y_linear)
```

```python
slope = model.coef_[0][0]
intercept = model.intercept_[0]
```

```python
print(slope)
print(intercept)
```

```python
new_data = np.array([[35],[45],[55]])
new_data = MinMaxScaler().transform(new_data)
predicted_data_new = model.predict(new_data)
predict_prices = MinMaxScaler().inverse_transform(predicted_data_new)
```

```python

plt.scatter(X_linear, y_linear, label ='area and price')
plt.plot(X_linear, model.predict(X_linear),color ='red', label = 'Linear Regression')
plt.scatter(new_data, predict_prices,color ='green',marker ='x',label ='predict new data')
plt.xlabel('Area')
plt.ylabel('Prices')
plt.legend()
plt.show()
```
